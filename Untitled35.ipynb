Question 1: What is the difference between K-Means and Hierarchical Clustering?
Provide a use case for each.

K-Means Clustering

Type: Partition-based clustering.

How it works:

You predefine the number of clusters k.

Algorithm assigns each data point to the nearest cluster centroid.

Iteratively updates centroids until convergence.

Strengths: Fast, works well on large datasets with spherical clusters.

Weaknesses: Requires k in advance, sensitive to outliers, struggles with non-spherical clusters.

✅ Use Case (K-Means):

Customer segmentation in retail/e-commerce.
Example: Grouping shoppers into clusters based on spending behavior (budget buyers, premium buyers, seasonal shoppers).


Hierarchical Clustering

Type: Tree-based clustering (dendrogram).

How it works:

Agglomerative (bottom-up): Start with each data point as a cluster and merge iteratively.

Divisive (top-down): Start with all data as one cluster and split iteratively.

No need to predefine number of clusters (you can "cut" the dendrogram at desired level).

Strengths: Produces hierarchy, good for small/medium datasets, interpretable.

Weaknesses: Computationally expensive for large datasets, sensitive to noise.

✅ Use Case (Hierarchical Clustering):

Document clustering in text mining.
Example: Grouping research papers into topics/subtopics while preserving a hierarchy (e.g., "AI → Machine Learning → Deep Learning").

Question 3: What are the core parameters of DBSCAN, and how do they influence the
clustering process?
Answer:

DBSCAN Core Parameters

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) groups together points that are closely packed and labels outliers as noise. It has two key parameters:

1. eps (ε – Epsilon)

Definition: The maximum distance between two points for them to be considered as neighbors.

Influence:

Small eps → Many small clusters, more points labeled as noise.

Large eps → Fewer, larger clusters (may merge distinct clusters).

2. min_samples (Minimum Samples)

Definition: The minimum number of points required to form a dense region (i.e., a cluster).

Influence:

Small min_samples → More clusters, even sparse regions may form clusters.

Large min_samples → Fewer clusters, more points considered as noise.

Question 4: Why is feature scaling important when applying clustering algorithms like
K-Means and DBSCAN?

Answer:

Feature scaling is important in clustering algorithms like K-Means and DBSCAN because these methods are distance-based. They calculate similarity between points using measures like Euclidean distance.

If features are on different scales (e.g., “Income” in lakhs vs. “Age” in years), the feature with the larger numerical range will dominate the distance computation.

This leads to biased clusters that do not represent the true structure of the data.

Effect on algorithms:

K-Means → Centroid calculation and cluster assignment get skewed toward high-scale features.

DBSCAN → The neighborhood radius (eps) becomes inconsistent across dimensions, affecting cluster density detection and outlier identification.

Solution:
Apply scaling techniques like:

Standardization (Z-score): Mean = 0, Std Dev = 1.

Normalization (Min-Max): Range scaled to [0, 1].

✅ Example:
In customer segmentation, without scaling, “Annual Income” may overshadow “Age.” After scaling, both features contribute fairly, leading to balanced and meaningful clusters.

Question 5: What is the Elbow Method in K-Means clustering and how does it help
determine the optimal number of clusters?

Answer:
The Elbow Method is a technique used in K-Means clustering to find the optimal number of clusters (k).

It works by plotting the Within-Cluster Sum of Squares (WCSS) or inertia against different values of k.

As k increases, WCSS decreases (clusters fit data better).

Initially, WCSS drops sharply, but after a certain point, the improvement becomes marginal — the curve looks like an “elbow.”

The “elbow point” represents the optimal number of clusters, balancing model simplicity and accuracy.

Question 6: Generate synthetic data using make_blobs(n_samples=300, centers=4),
apply KMeans clustering, and visualize the results with cluster centers.
(Include your Python code and output in the code box below.)

Answer:

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# Plot results with cluster centers
plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=30, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', s=200, alpha=0.75, marker='X', label='Centroids')
plt.title("KMeans Clustering with Cluster Centers")
plt.legend()
plt.show()
```

Question 7: Load the Wine dataset, apply StandardScaler , and then train a DBSCAN
model. Print the number of clusters found (excluding noise).
(Include your Python code and output in the code box below.)

```python
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import numpy as np

# Load wine dataset
wine = load_wine()
X = wine.data

# Apply StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train DBSCAN model
dbscan = DBSCAN(eps=1.8, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Count clusters (excluding noise labeled as -1)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
print("Number of clusters found (excluding noise):", n_clusters)

```

Question 8: Generate moon-shaped synthetic data using
make_moons(n_samples=200, noise=0.1), apply DBSCAN, and highlight the outliers in
the plot.
(Include your Python code and output in the code box below.)
Answer:

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# Generate moon-shaped synthetic data
X, y = make_moons(n_samples=200, noise=0.1, random_state=42)

# Apply DBSCAN
dbscan = DBSCAN(eps=0.2, min_samples=5)
labels = dbscan.fit_predict(X)

# Identify outliers (labeled as -1)
outliers = labels == -1

# Plot clusters and outliers
plt.figure(figsize=(8,6))
plt.scatter(X[~outliers, 0], X[~outliers, 1], c=labels[~outliers], cmap='viridis', s=30, label="Clusters")
plt.scatter(X[outliers, 0], X[outliers, 1], c='red', s=50, marker='x', label="Outliers")
plt.title("DBSCAN on make_moons (Outliers in Red)")
plt.legend()
plt.show()
```
